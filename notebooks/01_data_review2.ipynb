#### Create Daily Bed Occupancy Dataset (TARGET)
##### Your bed table is hourly, so aggregate it.

```python
beds_daily = (
    beds
    .groupby(
        [
            "hospital_id",
            "ward",
            "bed_type",
            "date"
        ]
    )
    .agg(
        total_beds=("total_beds","first"),
        staffed_beds=("staffed_beds","first"),
        occupied_beds=("occupied_beds","mean"),
        closed_beds=("closed_beds","mean")
    )
    .reset_index()
)
```

##### Create Occupancy Features

```python
beds_daily["occupancy_rate"] = (
    beds_daily["occupied_beds"] /
    beds_daily["staffed_beds"]
)
```

```python
beds_daily["available_beds"] = (
    beds_daily["staffed_beds"] -
    beds_daily["occupied_beds"]
)
```

#### Admissions Features
##### Daily admissions count

```python
admission_features = (
    admissions
    .groupby(
        [
            "hospital_id",
            "ward",
            "date"
        ]
    )
    .agg(
        daily_admissions=("admission_id","count"),
        avg_los_hours=("length_of_stay_hours","mean")
    )
    .reset_index()
)
```

##### Emergency Admissions

```python
emergency_features = (
    admissions[
        admissions["admission_type"]=="Emergency"
    ]
    .groupby(
        [
            "hospital_id",
            "ward",
            "date"
        ]
    )
    .size()
    .reset_index(
        name="emergency_admissions"
    )
)
```

#### ED Arrival Features

```python
ed_features = (
    arrivals
    .groupby(
        [
            "hospital_id",
            "date"
        ]
    )
    .agg(
        daily_ed_arrivals=("arrival_id","count")
    )
    .reset_index()
)
```

#### Surgery Demand Features

```python
surgery_features = (
    surgery
    .groupby(
        [
            "hospital_id",
            "date"
        ]
    )
    .agg(
        scheduled_surgeries=("surgery_id","count")
    )
    .reset_index()
)
```

#### Staffing Features
##### Staff has multiple roles, so aggregate.

```python
staff_features = (
    staff
    .groupby(
        [
            "hospital_id",
            "ward",
            "date"
        ]
    )
    .agg(
        planned_staff=("planned_staff","sum"),
        actual_staff=("actual_staff","sum")
    )
    .reset_index()
)
```

##### create staffing ratio

```python
staff_features["staffing_ratio"] = (
    staff_features["actual_staff"] /
    staff_features["planned_staff"]
)
```

#### Merge Everything

```python
model_df = beds_daily.copy()
```

##### Add Admissions

```python
model_df = model_df.merge(
    admission_features,
    on=[
        "hospital_id",
        "ward",
        "date"
    ],
    how="left"
)
```

##### Add Emergency Admissions

```python
model_df = model_df.merge(
    emergency_features,
    on=[
        "hospital_id",
        "ward",
        "date"
    ],
    how="left"
)
```

##### Add Staffing

##### Add Surgery

##### Add Time-Series Features

##### Lag 1 - Yesterday's occupancy

```python
model_df["lag_1_occupancy"] = (
    model_df
    .groupby(
        [
            "hospital_id",
            "ward"
        ]
    )
    ["occupancy_rate_model"]
    .shift(1)
)
```

##### Lag 7 - One week ago (Previous 7 days)

```python
model_df["lag_7_occupancy"] = (
    model_df
    .groupby(
        [
            "hospital_id",
            "ward"
        ]
    )
    ["occupancy_rate_model"]
    .shift(7)
)
```

##### 7-day moving average

```python
model_df["rolling_7_day_avg"] = (
    model_df
    .groupby(
        [
            "hospital_id",
            "ward"
        ]
    )
    ["occupancy_rate_model"]
    .transform(
        lambda x:
        x.rolling(7).mean()
    )
)
```

##### Day of week

```python
model_df["day_of_week"] = (
    model_df["date"]
    .dt.dayofweek
)
```

##### Day-to-Day Change

```python
model_df["occupancy_change"] = (
    model_df.groupby(
        [
            "hospital_id",
            "ward"
        ]
    )["occupancy_rate"]
    .diff()
)
```

##### Fill Lag NA Values

```python
lag_columns = [
    "lag_1_occupancy",
    "lag_7_occupancy",
    "rolling_7_day_avg",
    "occupancy_change"
]

model_df[lag_columns] = (
    model_df[lag_columns]
    .fillna(0)
)
```

#### Feature Engineering

##### Calendar Features

```python
model_df["date"] = pd.to_datetime(model_df["date"])

model_df["year"] = model_df["date"].dt.year

model_df["month"] = model_df["date"].dt.month

model_df["day"] = model_df["date"].dt.day

model_df["day_of_week"] = model_df["date"].dt.dayofweek

model_df["week_of_year"] = model_df["date"].dt.isocalendar().week.astype(int)

model_df["quarter"] = model_df["date"].dt.quarter

model_df["is_weekend"] = (
    model_df["day_of_week"] >= 5
).astype(int)
```

##### Bed Pressure

```python
model_df["bed_pressure"] = (
    model_df["occupied_beds"]
    -
    model_df["staffed_beds"]
)
```

##### Available Bed Percentage

```python
model_df["available_bed_pct"] = (
    model_df["available_beds"]
    /
    model_df["staffed_beds"]
)
```

##### Staff Shortage

```python
model_df["staff_shortage"] = (
    model_df["planned_staff"]
    -
    model_df["actual_staff"]
)
```

##### Emergency Admission Ratio

```python
model_df["emergency_ratio"] = (
    model_df["emergency_admissions"]
    /
    model_df["daily_admissions"]
)
```

```python
model_df["emergency_ratio"] = (
    model_df["emergency_ratio"]
    .fillna(0)
)
```

##### Surgery Pressure

```python
model_df["surgery_per_staff"] = (
    model_df["scheduled_surgeries"]
    /
    model_df["actual_staff"]
)

model_df["surgery_per_staff"] = (
    model_df["surgery_per_staff"]
    .fillna(0)
)
```
