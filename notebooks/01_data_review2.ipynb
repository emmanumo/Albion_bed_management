#### Standardize Columns and Texts

```python
def clean_columns(data):

    data.columns = (
        data.columns
        .str.lower()
        .str.strip()
        .str.replace(" ","_",regex=False)
    )

    return data
```

#### Create Daily Bed Occupancy Dataset (TARGET)
##### The bed table is hourly, so we aggregate to daily.

```python
staff_changes = (
    beds.groupby(
        ["hospital_id", "ward", "date"]
    )["staffed_beds"]
    .nunique()
)

print(staff_changes.value_counts())
```

```python
beds_daily = (
    beds
    .groupby(
        [
            "hospital_id",
            "ward",
            "bed_type",
            "date"
        ]
    )
    .agg(
        total_beds=("total_beds", "first"),
        staffed_beds=("staffed_beds", "mean"),
        occupied_beds=("occupied_beds", "mean"),
        closed_beds=("closed_beds", "mean")
    )
    .reset_index()
)
```

```python
beds_daily[
    [
        "total_beds",
        "staffed_beds",
        "occupied_beds",
        "closed_beds"
    ]
] = (
    beds_daily[
        [
            "total_beds",
            "staffed_beds",
            "occupied_beds",
            "closed_beds"
        ]
    ]
    .round(0)
    .astype(int)
)
```

```python
beds_daily.head()
```

##### Create Features

```python
beds_daily["occupancy_rate"] = (
    beds_daily["occupied_beds"] /
    beds_daily["staffed_beds"]
).round(1)
```

```python
beds_daily["available_beds"] = (
    beds_daily["staffed_beds"] -
    beds_daily["occupied_beds"]
).round(1)
```

#### Admissions Features
##### Daily admissions count

```python
admission_features = (
    admissions
    .groupby(
        [
            "hospital_id",
            "ward",
            "date"
        ]
    )
    .agg(
        daily_admissions=("admission_id","count"),
        avg_los_hours=("lenth_of_stay_hours","mean")
    )
    .reset_index()
)
```

```python
admission_features[
    [
        "daily_admissions",
        "avg_los_hours"
    ]
] = (
    admission_features[
        [
            "daily_admissions",
            "avg_los_hours"
        ]
    ]
    .round(0)
    .astype(int)
)
```

```python
admission_features.head()
```

##### Emergency Admissions

```python
emergency_features = (
    admissions[
        admissions["admission_type"]=="Emergency"
    ]
    .groupby(
        [
            "hospital_id",
            "ward",
            "date"
        ]
    )
    .size()
    .reset_index(
        name="emergency_admissions"
    )
)
```

#### ED Arrival Features

```python
ed_features = (
    arrivals
    .groupby(
        [
            "hospital_id",
            "date"
        ]
    )
    .agg(
        daily_ed_arrivals=("arrival_id","count")
    )
    .reset_index()
)
```

#### Surgery Demand Features

```python
surgery_features = (
    surgery
    .groupby(
        [
            "hospital_id",
            "date"
        ]
    )
    .agg(
        scheduled_surgeries=("surgery_id","count")
    )
    .reset_index()
)
```

#### Staffing Features
##### Staff has multiple rows, so we aggregate.

```python
staff_features = (
    staff
    .groupby(
        [
            "hospital_id",
            "ward",
            "date"
        ]
    )
    .agg(
        planned_staff=("planned_staff","sum"),
        actual_staff=("actual_staff","sum")
    )
    .reset_index()
)
```

##### create staffing ratio

```python
staff_features["staffing_ratio"] = (
    staff_features["actual_staff"] /
    staff_features["planned_staff"]
).round(1)
```

#### Merge Everything

```python
model_df = beds_daily.copy()
```

##### Add Admissions

```python
model_df = model_df.merge(
    admission_features,
    on=[
        "hospital_id",
        "ward",
        "date"
    ],
    how="left"
)
```

##### Add Emergency Admissions

```python
model_df = model_df.merge(
    emergency_features,
    on=[
        "hospital_id",
        "ward",
        "date"
    ],
    how="left"
)
```

##### Add Staffing

```python
model_df = model_df.merge(
    staff_features,
    on=[
        "hospital_id",
        "ward",
        "date"
    ],
    how="left"
)
```

##### Add ED Arrivals

```python
model_df = model_df.merge(
    ed_features,
    on=[
        "hospital_id",
        "date"
    ],
    how="left"
)
```

##### Add Surgery

```python
model_df = model_df.merge(
    surgery_features,
    on=[
        "hospital_id",
        "date"
    ],
    how="left"
)
```

##### Add Hospital Information

```python
model_df = model_df.merge(
    hospital,
    on="hospital_id",
    how="left"
)
```
